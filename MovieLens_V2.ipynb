### Upload dataset Movie List

```
from google.colab import files
uploaded = files.upload()
```

```
import io
movie_list = pd.read_csv(io.BytesIO(uploaded['movies.csv']))
```

```
movie = movie_list.loc[:,["movieId","title"]]
```

```
movie
```

```
dataset = pd.merge(movie,rating)
```

```
dataset.shape
```

```
daset_pivot = dataset.pivot_table(index=["userId"], columns=["title"], values="rating")
```

```
daset_pivot.head(10)
```

```
movie_watched = daset_pivot["10 Cloverfield Lane (2016)"]
similarity_with_other_movies = daset_pivot.corrwith(movie_watched)
similarity_with_other_movies = similarity_with_other_movies.sort_values(ascending=False)

similarity_with_other_movies.head(10)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

vect = TfidfVectorizer()
tfidf_matrix = vect.fit_transform(dataset)
```

```
table_dataset = pd.DataFrame(tfidf_matrix.toarray(), columns = vect.get_feature_names())
```

```
table_dataset.sort_values(by="rating", ascending=False)
```
